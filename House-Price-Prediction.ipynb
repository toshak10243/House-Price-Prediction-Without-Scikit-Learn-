```python
# Step 1: Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create Sample Data (House Size & Price)
data = {
    "Size (sq ft)": [750, 800, 850, 900, 950, 1000, 1050, 1100, 1150, 1200],
    "Price (₹ in Lakhs)": [35, 38, 40, 43, 45, 48, 50, 53, 55, 58]
}

df = pd.DataFrame(data)

# Step 3: Extract Features (X) and Target (y)
X = df["Size (sq ft)"].values
y = df["Price (₹ in Lakhs)"].values

# Step 4: Compute Mean of X and y
X_mean = np.mean(X)
y_mean = np.mean(y)

# Step 5: Calculate Slope (m) and Intercept (b) for y = mx + b
m = np.sum((X - X_mean) * (y - y_mean)) / np.sum((X - X_mean) ** 2)
b = y_mean - (m * X_mean)

# Step 6: Predict Values
y_pred = m * X + b

# Step 7: Plot the Data and Regression Line
plt.scatter(X, y, color="blue", label="Actual Prices")
plt.plot(X, y_pred, color="red", label="Predicted Prices (Regression Line)")
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price (₹ in Lakhs)")
plt.title("House Price Prediction using Linear Regression")
plt.legend()
plt.show()

# Step 8: Predict Price for New House Size
new_size = 1100  # Change this value to test predictions
predicted_price = m * new_size + b
print(f"Predicted Price for {new_size} sq ft house: ₹{predicted_price:.2f} Lakhs")

```
